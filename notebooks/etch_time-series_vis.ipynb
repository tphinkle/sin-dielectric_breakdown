## Imports

```python
# Python standard library
import csv

# Scientific computing
import numpy as np
import matplotlib.pyplot as plt
```

```python
rho = 1
L = 50*10**(-9.)
A = np.pi*(1*10**(-6.))**2.
print rho*L/A

```

## Load data

```python
file_path = '../data/SiN-10nm-8V-pH10b_35min.txt'
```

```python
times = []
currents = []

with open(file_path, 'r') as file_handle:
    file_reader = csv.reader(file_handle, delimiter = '\t')
    for row in file_reader:
        times.append(float(row[0]))
        currents.append(float(row[1]))
        
times = np.array(times)
currents = np.array(currents)
```

## Calculate radii

```python
L = 10*10**(-9.)
rho = 0.1
V = 8

radii = np.sqrt(currents*rho*L/(np.pi*V))
```

```python
plt.plot(times, radii)
plt.show()
```

## Plots

```python
fig, axes = plt.subplots(1,2, figsize = (8,3))

# Current plot
plt.sca(axes[0])
plt.semilogy(times, currents)

plt.xlabel('Time (s)')
plt.ylabel('Current (A)')
plt.grid()

# Radii plot
plt.sca(axes[1])
plt.semilogy(times, radii)

plt.xlabel('Time (s)')
plt.ylabel('Radius (m)')
plt.grid()

fig.tight_layout()

plt.show()
```

## Plot adjusted

```python
leakage_current = currents[np.where(times == 1700)[0]]
adjusted_currents = currents - leakage_current


L = 10*10**(-9.)
rho = 1/10.862
V = 8

adjusted_radii = np.sqrt(adjusted_currents*rho*L/(np.pi*V))
```

```python
fig, axes = plt.subplots(1,2, figsize = (8,3))

# Current plot
plt.sca(axes[0])
plt.semilogy(times, adjusted_currents)

plt.xlabel('Time (s)')
plt.ylabel('Current (A)')
plt.grid()

# Radii plot
plt.sca(axes[1])
plt.semilogy(times, adjusted_radii)

plt.xlabel('Time (s)')
plt.ylabel('Radius (m)')
plt.grid()

for i in range(2):
    axes[i].set_xlim(5400,5750)
#axes[0].set_ylim(.6*10**-7., 10**-6.)
#axes[1].set_ylim(10**(-9.), 10**(-8.))

fig.tight_layout()

plt.show()
```
